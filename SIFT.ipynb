```python
import cv2
```

```python
sift=cv2.SIFT_create()
```

```python
#block feature matching
bf=cv2.BFMatcher(cv2.NORM_L2,crossCheck=True)
```

```python
img1=cv2.imread(r'C:\Users\HP\OneDrive\Desktop\LieDetector\LieDetection\data1\st17.jpg')
img2=cv2.imread(r'C:\Users\HP\OneDrive\Desktop\LieDetector\LieDetection\data1\st24.jpg')
```

```python
#converting rgb image to grayscale
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
```

```python
keypoints_1,descriptors_1=sift.detectAndCompute(img1,None)
keypoints_2,descriptors_2=sift.detectAndCompute(img2,None)
```

```python
matches=bf.match(descriptors_1,descriptors_2)
matches=sorted(matches,key=lambda x:x.distance)
```

```python
img3=cv2.drawMatches(img1,keypoints_1,img2,keypoints_2,matches[300:600]
                    ,img2,flags = 2 )
```

```python
cv2.imshow('SIFT',img3)
cv2.waitKey(0)
```

```python
#SIFT for feature extraction
import matplotlib.pyplot as plt
import cv2
 
# Loading the image
img = cv2.imread(r'C:\Users\HP\OneDrive\Desktop\LieDetector\LieDetection\st16.jpg')
 
 # Converting image to grayscale
gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
 
# Applying SIFT detector
sift = cv2.xfeatures2d.SIFT_create()
kp = sift.detect(gray, None)
 
# Marking the keypoint on the image using circles
img=cv2.drawKeypoints(gray ,
                      kp ,
                      img ,
                      flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
 
cv2.imwrite('image-with-keypoints.jpg', img)
plt.imshow(img)
```
